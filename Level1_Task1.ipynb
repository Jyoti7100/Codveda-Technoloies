```python
# import data 
data = pd.read_csv("2) Stock Prices Data Set.csv", encoding='unicode_escape')
```

```python
print(data.isnull().sum())
```

```python
print("Total missing values:",data.isnull().sum().sum())
```

```python
numeric_cols=['open','high','low']
for col in numeric_cols:
    data[col]=data[col].fillna(data[col].median())
```

```python
print(data.isnull().sum())
print("Total missing values:", data.isnull().sum().sum())
```

```python
data=data.dropna()
```

```python
print(data.shape)
```

```python
print('Duplicate rows:',data.duplicated().sum())
```

```python
data=data.drop_duplicates()
print(data.shape)
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
print(data.dtypes)
```

```python
data['symbol'] = data['symbol'].str.upper().str.strip()
```

```python
data.to_csv("Cleaned_Stock_Prices.csv", index=False)
```

```python
# Check duplicate rows
print("Duplicate rows:", data.duplicated().sum())

# Remove duplicate rows
data = data.drop_duplicates()

# Verify new shape
print("Shape after removing duplicates:", data.shape)
```

The dataset contained 27 missing values in the open, high, and low columns.

Missing values were imputed using the median of each respective column.

After imputation, the dataset contained no missing values.

Applying dropna() did not remove any rows because all missing values had already been handled.

The cleaned dataset retained all 497,472 rows and 7 columns.
